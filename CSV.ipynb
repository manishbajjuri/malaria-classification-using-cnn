```python
import cv2
import os
import numpy as np
import csv
import glob
import time
start = time.time()
```

# Preparing Header lables 

```python
header =["label"]

for i in range(0,10000):
        header.append("Pixel  "+str(i))

with open('train.csv', 'a') as f:
        writer = csv.writer(f)
        writer.writerow(header)

with open('test.csv', 'a') as f:
        writer = csv.writer(f)
        writer.writerow(header)
```

# Train Prasitized 

```python
label = "Parasitized"
dirList = glob.glob("data/train/"+label+"/*.png")
for img_path in dirList:
	im = cv2.imread(img_path)
	im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
	#im_gray = cv2.GaussianBlur(im_gray, (5, 5), 0)
	roi = cv2.resize(im_gray, (100, 100), interpolation=cv2.INTER_AREA)
	data=[]
	data.append(label)
	rows,cols = roi.shape
	for i in range(rows):
	    for j in range(cols):
	        k = roi[i,j]       
	        data.append(k)
	with open('train.csv', 'a') as f:
		writer = csv.writer(f)
		writer.writerow(data)
```
